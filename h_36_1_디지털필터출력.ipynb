## 디지털 필터 출력
- 동일한 패턴일 때 가장 큰 값을 얻는다

```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
x_org = np.array([0.2, 0.7, 0.2, 0, -0.2, -0.7, -0.2])
plt.plot(x_org, 'o-')
print('최대값: ',(x_org*x_org).sum())
```

```python
np.random.seed(9)
for i in range(10):
    x = np.random.uniform(-1,1,7)
    plt.plot(x, 'o-')
    plt.plot(x_org, 'o-')
    plt.show()
    print((x*x_org).sum().round(2))
```
